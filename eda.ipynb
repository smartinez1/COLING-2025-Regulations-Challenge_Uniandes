```python
import pandas as pd

# Load the CSV file
csv_filename = 'recursive_data/QA/QA.csv'  # Adjust the path as necessary
df = pd.read_csv(csv_filename)
```

```python
# Display basic information about the DataFrame
print("Basic Information:")
print(df.info())

# Display the first few rows of the DataFrame
print("\nFirst 5 Rows:")
print(df.head())

# Display summary statistics
print("\nSummary Statistics:")
print(df.describe(include='all'))  # Include 'all' to get stats for non-numeric columns too

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Count the number of unique URLs and contents
unique_urls_count = df['url'].nunique()
unique_contents_count = df['content'].nunique()
print(f"\nUnique URLs: {unique_urls_count}")
print(f"Unique Contents: {unique_contents_count}")

# Show duplicates in the 'url' and 'content' columns, if any
print("\nDuplicate URLs:")
print(df[df.duplicated(subset='url', keep=False)])  # Show all duplicates

print("\nDuplicate Contents:")
print(df[df.duplicated(subset='content', keep=False)])  # Show all duplicates

# Optional: Display a sample of 5 random rows from the DataFrame
print("\nRandom Sample of 5 Rows:")
print(df.sample(5))
```
